#### We can use YAML files as well to store the configurations but Since, they are not the python classes hence, it is difficult to get auto suggestions from them and developer need to remember them always. Hence, we can either use configurations class as below or read the yaml file and store the things in config class. So that it can be developer friendly as well.

```python
class Config:
    def __init__(self):
        self.work_dir=workdir
        self.data_dir=os.path.join(self.work_dir,"data")
        self.csv_file_name="data.csv"
        self.csv_file_path = os.path.join(self.data_dir,self.csv_file_name)
        self.pipeline_root = os.path.join(os.getcwd(),"aiops_pipeline_artifacts")
        self.pipeline_name = "aiops_pipeline"
        self.meta_data_path = os.path.join("Aiops_pipeline_metadata", self.pipeline_name,"metadata.db")     ## Here we would like to create a metadata file which will be stored in sqlite        
```

```python
config = Config()
```

```python
from tfx.orchestration.experimental.interactive.interactive_context import InteractiveContext
```

```python
interactive_context= InteractiveContext()
```

```python
### Below we are going to use sqlite for our metadata construction but we can also use other databases, we have those classes available as well.
```

```python
from tfx.orchestration.metadata import sqlite_metadata_connection_config  
```

```python
## This will create a pipleline folder taking configs from our config class

interactive_context= InteractiveContext(

    pipeline_name=config.pipeline_name,
    pipeline_root=config.pipeline_root,
    metadata_connection_config=sqlite_metadata_connection_config(config.meta_data_path)
    

)
```

```python
## This library is used for converting csv file into tf record file
from tfx.components import CsvExampleGen
```

```python
## Here we have initialize the class of the csv_example with our input 
csv_example_gen = CsvExampleGen(input_base=config.data_dir)
```

```python
## Now we will make the final things with interactive_context
interactive_context.run(csv_example_gen)
```
